## Grouping Data

we are going to learn how to group data in a dataframe and perform operation through the following way

-  split the object and combine the result
-  Iterate the Group
-  View the Group
-  Perform Aggregation Operation on Groups




We are going to learn how to group data in a dataframe and perform operation through the following way:

- split the object and combine the result - This involves using groupby() to split data into groups based on some criteria         then recombining the results
- Iterate the Group - How to loop through each group to perform operations on them individually
- View the Group - Methods to examine the structure and content of grouped data
- Perform Aggregation Operation on Groups - Using functions like sum(), mean(), count() etc. on grouped data to get summary      statistics


```python
# import pandas and numpy

import pandas as pd
import numpy as np

# Dataset

dataset = {
    'player': ['ada', 'chima', 'obi', 'lyon', 'stacy','jcn'],
    'rank' : [1, 4, 3, 5, 2, 6],
    'year' : [2023, 2022, 2021, 2022, 2018, 2015]
}

# create a DataFrame using the DataFrame() method 

df = pd.DataFrame (dataset)

# Display the records

print("Cricket Player Records\n", df)

#Group the data on player value
val = df.groupby('player')

#Display the first entry
print("\n", val.first())
```

```python
# Iterate the Group - How to loop through each group to perform operations on them individually

# Dataset

dataset = {
    'player': ['ada', 'chima', 'obi', 'lyon', 'stacy','jcn'],
    'rank' : [1, 4, 3, 5, 2, 6],
    'year' : [2023, 2022, 2021, 2022, 2018, 2015]
}

# create a DataFrame using the DataFrame() method 

df = pd.DataFrame (dataset)

# Display the records

print("Cricket Player Records\n", df)

#Group the data on player value
val = df.groupby('player')

# Iterate

for name, group in val:
    print("\n", name)
    print(group)
```

```python
# View the Group - Methods to examine the structure and content of grouped data

# Dataset

dataset = {
    'player': ['ada', 'chima', 'obi', 'lyon', 'stacy','jcn'],
    'rank' : [1, 4, 3, 5, 2, 6],
    'year' : [2023, 2022, 2021, 2022, 2018, 2015]
}

# create a DataFrame using the DataFrame() method 

df = pd.DataFrame (dataset)

# Display the records

print("Cricket Player Records\n", df)

# Group by player and Display 

print("\nView Group:\n", df.groupby('player').groups)
```

```python
# Aggregation operation - Get the mean of the group data

# Import necessary libraries
import pandas as pd
import numpy as np

# Dataset
dataset = {
    'player': ['ada', 'chima', 'obi', 'lyon', 'stacy','jcn'],
    'rank' : [1, 4, 3, 5, 2, 6],
    'points': [95, 70, 65, 80, 90, 50],
    'year' : [2023, 2022, 2021, 2022, 2018, 2015]
}

# create a DataFrame using the DataFrame() method 
df = pd.DataFrame(dataset)

# Display the records
print("Cricket Player Records\n", df)

# use the groupby to group
groupRes = df.groupby('year')

# use the agg() as a method on the groupby object
print("\nmean:\n", groupRes['points'].agg(np.mean))
```

```python
# Perform Aggregation Operation on Groups

# Import necessary libraries
import pandas as pd
import numpy as np

# Dataset
dataset = {
    'player': ['ada', 'chima', 'obi', 'lyon', 'stacy','jcn'],
    'rank' : [1, 4, 3, 5, 2, 6],
    'points': [95, 70, 65, 80, 90, 50],
    'year' : [2023, 2022, 2021, 2022, 2018, 2015]
}

# create a DataFrame using the DataFrame() method 
df = pd.DataFrame(dataset)

# Display the records
print("Cricket Player Records\n", df)

# use the groupby to group
groupRes = df.groupby('year')

# the agg() performs aggregation
# the numpy.size attribute returns the size of each group

print("\n",groupRes.agg(np.size))
```
