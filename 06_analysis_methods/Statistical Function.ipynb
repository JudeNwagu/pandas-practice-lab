## Pandas Statistical Function

we are going to look at most useable satistical function in pandas:

- **SUM()** to get the sum of a value in our dataframe
- **Count()** we use the count method to get non-empty values
- **Max()** method in python pandas is used to return the largest or maximum number in our dataframe
- **Min()** method in python pandas is used to return the lowest or minimum number in our dataframe
- **Mean()** method in python pandas is used to return the lowest or minimum number in our dataframe
- **Median()** in python pandas is used to return the median values in our dataframe
- **std()** in python pandas is used to get the standard deviation value in our dataframe
- **Described()** in python pandas is used to get the summary statistics of our dataset or dataframe



```python
import pandas as pd
import numpy as np
```

```python
# using our first example which is sum
# Dataset creating our dataset

mark = {
    'language': [80, 90, 85, 83, 75],
    'Tech' : [78, 82, 75, 69, 72],
    'maths' : [95, 70, 80, 60, 90],
    'english' : [95, 84, 91, 75, 80],
    'science': [80, 91, 85, 94, 90],
    'agric': [89, 80, 79, 81, 83],
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (mark)

# print the student records

print("Student Records\n\n", df)

# display the sum each subject
print("\n Record Of Each Subject\n",df.sum())
```

```python
# Count(): we use the count method to get non-empty values
mark = {
    'language': [80, 90, 85, 83, None],
    'Tech' : [None, 82, 75, None, 72],
    'maths' : [95, 67, None, 60, 90],
    'english' : [67, 84, 91, 75, 40],
    'science': [80, None, 85, None, 90],
    'agric': [None, 80, None, None, 83],
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (mark)

# Display the Student record

print("Student Records\n\n", df)

# display the sum each subject
print("\nCount Number Of Non-Empty Value\n",df.count())
```

```python
# Max() method in python pandas is used to return the largest or maximum number in our dataframe


mark = {
    'language': [80, 90, 85, 83, 75],
    'Tech' : [78, 82, 75, 69, 72],
    'maths' : [95, 70, 80, 60, 90],
    'english' : [95, 84, 91, 75, 80],
    'science': [80, 91, 85, 94, 90],
    'agric': [89, 80, 79, 81, 83],
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (mark)

# print the student records

print("Student Records\n\n", df)

# display the sum each subject
print("\n Return The Max Num Of Each Subject\n",df.max())
```

```python
# Min() method in python pandas is used to return the lowest or minimum number in our dataframe


mark = {
    'language': [80, 90, 85, 83, 75],
    'Tech' : [78, 82, 75, 69, 72],
    'maths' : [95, 70, 80, 60, 90],
    'english' : [95, 84, 91, 75, 80],
    'science': [80, 91, 85, 94, 90],
    'agric': [89, 80, 79, 81, 83],
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (mark)

# print the student records

print("Student Records\n\n", df)

# display the sum each subject
print("\n Return The Minimum Num Of Each Subject\n", df.min())
```

```python
# Mean() method in python pandas is used to return the lowest or minimum number in our dataframe

mark = {
    'language': [80, 90, 85, 83, 75],
    'Tech' : [78, 82, 75, 69, 72],
    'maths' : [95, 70, 80, 60, 90],
    'english' : [95, 84, 91, 75, 80],
    'science': [80, 91, 85, 94, 90],
    'agric': [89, 80, 79, 81, 83],
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (mark)

# print the student records

print("Student Records\n\n", df)

# display the sum each subject
print("\n Average OR Mean Of Each Subject\n", df.mean())
```

```python
# Median() in python pandas is used to return the median values in our dataframe
mark = {
    'language': [80, 90, 85, 83, 75],
    'Tech' : [78, 82, 75, 69, 72],
    'maths' : [95, 70, 80, 60, 90],
    'english' : [95, 84, 91, 75, 80],
    'science': [80, 91, 85, 94, 90],
    'agric': [89, 80, 79, 81, 83],
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (mark)

# print the student records

print("Student Records\n\n", df)

# display the sum each subject
print("\n Return The Median Num Of Each Subject\n", df.median())
```
