# CATEGORICAL DATA


In this part we would learn how to work with categorical data in pandas. 
catgorical data is a data type corresponding to the categorical variables in statistics. 
A categorical variable takes on a limited number of pssible values. Examples  are gender, blood type, country, rating etc

using this two examples we would create two categories first

* create a categorical series in pandas
* create a categorical DataFarme in pandas




```python

# import pandas

import pandas as pd

# Create a Categorical series

s = pd.Series(["a", "e", "i", "o", "u"], dtype = "category")


print("Series = \n", s)

```

```python
# create a categorical dataframe using the dtype="category" while creating a dataframe for categorical Dataframe

#cre
df =pd.DataFrame ( {"cat1": list("pqrs"), "cat2": list("pqrs"), "cat3": list("pqrs")}, dtype = "category")

# Display the Dataframe

print("DataFrame = \n", df)

#Display the Datatypes
                  
print("\nData types of each colum = \n", df.dtypes)                  
```

```python
# APPEND new Categories in pandas using "cat.add"

#create a categorical series

s= pd.Series(["Liv", "Manu", "Ars", "Che", "Manc"],dtype = "category")

# Display the series

print("Current Series = \n", s)

# Append a category
s = s.cat.add_categories("Tot")

#Display the updated category 

print("\nUpdate category list = \n", s)
```

```python
# Remove a category using the cat.remove_categories() method

#create a categorical series

s= pd.Series(["Liv", "Manu", "Ars", "Che", "Manc"],dtype = "category")

# Display the series

print("Current Series = \n", s)

# Remove a category from the list
s = s.cat.remove_categories("Che")

# Display the updated category

print("\nUpdated new category =\n", s)
```
