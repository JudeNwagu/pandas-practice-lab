```python
# COMBINE TWO PANDAS SERIES

''' using combine in pandas to fetech the largest values from both series. 
Each element of both the series will be compared one by one
'''

# import pandas
import pandas as pd

# Data

data1 = [10, 20, 40, 80, 100]
data2 = [25, 5, 75, 95, 45]

# Create a series using the series () method

series1 = pd.Series(data1)
series2 = pd.Series(data2)

# Display the series

print("series1:\n", series1)
print("series2:\n", series2)

def demo(x1, x2):
    if(x1 > x2):
        return x1
    else:
        return x2
# combine the two series into one

res = series1.combine(series2, demo)

# Display the result
print("\nAfter Combining\n", res)
    


```
