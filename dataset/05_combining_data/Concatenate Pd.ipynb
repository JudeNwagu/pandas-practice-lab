```python
# concatenate two pandas DataFrame

import pandas as pd

# Dataset

dataset1 = {
    'Id': ["ST1", "ST2", "ST3", "ST4", "ST5",],
    'Student' : ['ada', 'chima', 'obi', 'lyon', 'stacy'],
    'Roll' : [101, 102, 103, 104, 105]

}

dataset2 = {
    'Id' : ["ST6", "ST7 ","ST8", "ST9"],
    'Student' : ['chisom','mimi','coco','teni'],
    'Roll' : [106, 107, 108, 109]

}

# DataFrames
dataframe1 = pd.DataFrame(dataset1, index =["Student1", "Student2", "Student3", "Student4", "Student5",])
print("DataFrame1 = \n", dataframe1)

dataframe2 = pd.DataFrame(dataset2, index = ["Student6", "Student7", "Student8", "Student9"])
print("DataFrame2 = \n", dataframe2)

# concatenate two DataFrames
catDF = pd.concat([dataframe1, dataframe2])
print("\nconcatenating DataFrames = \n", catDF)
```
