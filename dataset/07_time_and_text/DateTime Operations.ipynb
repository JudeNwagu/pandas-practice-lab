```python
# Get the current date and time using the Timestamp.now()

# import pandas

import pandas as pd

print("\nCurrent Date and Time:\n", pd.Timestamp.now())
```

```python
# Get the day of the week using dayofweek()

# construct a timetstamp
timestamp = pd.Timestamp(year= 2026, month=2, day=19, hour=8)

# Display the date
print("Date and Time\n", timestamp)

# Display day of the week
print("\nDay Of Week",timestamp.dayofweek)
```

```python
# Get the day of the week using dayofweek()


# Provide a date to the Timestamp constructor
# Format can be 'YYYY-MM-DD' or you can provide year, month, day as separate arguments
timeStamp = pd.Timestamp('2023-11-15')  # Using a string date

# Alternative ways to create a timestamp:
# timeStamp = pd.Timestamp(year=2023, month=11, day=15)  # Using separate arguments
# timeStamp = pd.Timestamp(2023, 11, 15)  # Using positional arguments

# Now you can use dayofweek
day_of_week = timeStamp.dayofweek
print(f"Day of week: {day_of_week}")  # 0=Monday, 1=Tuesday, ..., 6=Sunday
```

```python
# Get the day of the year

# Provide a date to the Timestamp constructor
# Format can be 'YYYY-MM-DD' or you can provide year, month, day as separate arguments
timeStamp = pd.Timestamp('2025-11-15')  # Using a string date

# Alternative ways to create a timestamp:
# timeStamp = pd.Timestamp(year=2023, month=11, day=15)  # Using separate arguments
# timeStamp = pd.Timestamp(2023, 11, 15)  # Using positional arguments

# Display day of year
print("\nDay Of year:\n",timeStamp.dayofyear)
```

```python
# Get the number of days in a month


# construct a timetstamp
timestamp = pd.Timestamp(year= 2026, month=2, day=19, hour=8)

# Display the date
print("Date and Time\n", timestamp)

# Display number of days in the month
print("\nDays in month:\n", timestamp.daysinmonth)
```

```python
# check if the year is a leap year

# construct a timetstamp
timestamp = pd.Timestamp(year= 2026, month=2, day=19, hour=8)

# Display the date
print("Date and Time\n", timestamp)

# check for leap year

print("\nCheck if its leap year:\n", timestamp.is_leap_year)
```

```python
# check if the date is the last day of the month

# construct a timetstamp
timestamp = pd.Timestamp(year= 2026, month=2, day=28, hour=8)

# Display the date
print("Date and Time\n", timestamp)

# check if the date is the end of month
print("\nIs this the month end?\n", timestamp.is_month_end)
```

```python
# check if the date is the first dayy of the month

# construct a timetstamp
timestamp = pd.Timestamp(year= 2026, month=2, day=28, hour=8)

# Display the date
print("Date and Time\n", timestamp)


# check if the date is the first day of year
print("\nIs this the first day of the month?\n", timestamp.is_month_start)
```

```python
# check if the date is the last day of the year

# construct a timetstamp
timestamp = pd.Timestamp(year= 2026, month=2, day=28, hour=8)

# Display the date
print("Date and Time\n", timestamp)

# check if the date is the last day of the year
print("\nIs this the last day of the year?\n",timestamp.is_year_end)
```

```python
#check if the date is the first day of the year

# construct a timetstamp
timestamp = pd.Timestamp(year= 2026, month=2, day=28, hour=8)

# Display the date
print("Date and Time\n", timestamp)

# check if the date is the first day of the year
print("\nIs this the first day of the year?\n", timestamp.is_year_start)
```
