```python
# PANDAS SERIES
'''
- Create a Pandas series
- Accessa value from a pandas series
- Name your own indexes in a pandas series
- Access a value from a pandas series with labels
'''
# import pandas

import pandas as pd
 
# Data
data = [10, 20, 40, 80, 100]

# Create a pandas series
s = pd.Series(data)
```

```python
# Accessing a value from a pandas series

'''
The [] is used to access a value. set the index of the value you want to display inside []:
'''

# Data
data = [10, 20, 40, 80, 100]

# Create a pandas series
s = pd.Series(data)

print("Series:\n", s)

# Access a value
print("\nValue from a Pandas Series:", s[3])
```

```python
# Name your indexes in a pandas series

'''
The index arguments is used to set and name your own indexes in a series 
i.e the labels can be set accordingly

'''
# Data

data = [10, 20, 40, 80, 100]

# create a series using the series() method

s = pd.Series(data, index = ["RowA", "RowB", "RowC", "RowD", "RowE"])

# Display the series
print("\nSeries (with a custom index labels)\n" ,s)
```

```python
# Access a value from a pandas series with labels

# Data

data = [10, 20, 40, 80, 100]

# create a series using the series() method

s = pd.Series(data, index = ["RowA", "RowB", "RowC", "RowD", "RowE"])

# Display the series
print("\nSeries (with a custom index labels)\n" ,s)

# Access a value referring the label

print("\nValues from a Pandas Series\n", s["RowC"])
```
