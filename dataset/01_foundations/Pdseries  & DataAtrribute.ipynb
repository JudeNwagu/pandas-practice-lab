```python
# pandas dtype attribute

# import pandas as pd
import pandas as pd
# Data
data = [10, 20, 40, 60, 80, 100]

# create a series using the series() method
s = pd.Series(data)

# Display the series
print("series:\n", s)

# Data type
print("\nSeries Datatype:", s.dtype)
```

```python
# Ndim Dimension

# Data
data = [10, 20, 40, 60, 80, 100]

# create a series using the series() method
s = pd.Series(data)

# Display the series
print("series:\n", s)

# Dimension
print("\nSeries Diemension:\n" , s.ndim)
```

```python
# Pandas Size attribute
# The pandas.series.size is used to return the number of elements in the pandas series

# Data
data = [10, 20, 40, 60, 80, 100]

# create a series using the series() method
s = pd.Series(data)

# Display the series
print("series:\n", s)

# display the size

print("\nSeries size:\n", s.size)
```

```python
# Name attributes
# The pandas.series.name is used to return the name of the series in pandas


# Data
data = [10, 20, 40, 60, 80, 100]

# create a series using the series() method
s = pd.Series(data, name = "My Number Series")

# Display the series
print("series:\n", s)

# Return the name of the attribute

print("\nSeries name:\n", s.name)
```

```python
# HASNANS ATTRIBUTE

import numpy as np

# Data
data = [10, 20, 40, 60, 80, 100, np.nan]

# create a series using the series() method
s = pd.Series(data, name = "My Number Series")

# Display the series
print("series:\n", s)

print("\nDoes the Series has NAN:\n", s.hasnans)
```

```python
# Index attribute
# the pandas.series attribute is used to display the index of the pandas series

# Data
data = [10, 20, 40, 80, 100,]

# create a series using the series() method
s = pd.Series(data, index =['Num1', 'Num2', 'Num3', 'Num4', 'Num5'], name = "My Number Series")
 
# Display the series
print("series:\n", s)

# Return the index

print("\nSeries Index:\n",s.index)
```

```python
# Head() method
# The pandas.series.head() method is used to return the first n rows of the pandas

# Data
data = [10, 20, 40, 80, 100, 120, 130, 160, 180, 200]

# create a series using the series() method
s = pd.Series(data, name = "My Number Series")
 
# Display the series
print("series:\n", s)

# display the first five row

print("\nThe first five rows:\n", s.head())
print("\nThe first two rows:\n", s.head(2))
```

```python
# Pandas Series Tail() method

# The pandas.series.tail() method is used to return the last n rows of the pandas

# Data
data = [10, 20, 40, 80, 100, 120, 130, 160, 180, 200]

# create a series using the series() method
s = pd.Series(data, name = "My Number Series")
 
# Display the series
print("series:\n", s)

# display the first five row

print("\nThe last five rows:\n", s.tail())
print("\nThe last two rows:\n", s.tail(2))
```

```python
# INFO() method

#The pandas.series.info() method is used to display the summary of the pandas

# Data
data = [10, 20, 40, 80, 100,]

# create a series using the series() method
s = pd.Series(data, index =['Num1', 'Num2', 'Num3', 'Num4', 'Num5'], name = "My Number Series")
 
# Display the series
print("series:\n", s)

#show the summary of the data

print("\nSeries Summary:\n", s.info())
```
