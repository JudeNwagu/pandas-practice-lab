```python
#pandas DataFrame dtypes attribute

# import pandas

import pandas as pd

# Dataset

dataset = {
    'Student': ['ada', 'chima', 'obi', 'lyon', 'stacy'],
    'rank' : [1, 4, 3, 5, 2],
    'mark' : [95, 70, 80, 60, 90]
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (dataset, index =['RowA', 'RowB', 'RowC', 'RowD', 'RowE'])

# print the student records

print("Student Records\n\n", df)

# print the dataframe dtypes

print("\nDatatypes:\n", df.dtypes)
```

```python
# PANDAS DATAFRAME - NDIM
# The pandas.DataFrame.ndim is used to return the number of dimensions of the DataFrame

# Dataset

dataset = {
    'Student': ['ada', 'chima', 'obi', 'lyon', 'stacy'],
        'rank' : [1, 4, 3, 5, 2],
    'mark' : [95, 70, 80, 60, 90]
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (dataset, index =['RowA', 'RowB', 'RowC', 'RowD', 'RowE'])

# print the student records

print("Student Records\n\n", df)

# Numbers of dimension int the DataFrame

print("\nNumber of Dimensions:\n",df.ndim)

```

```python
# The pandas DataFrame.size is used to return the number of elements in the DataFrame

# Dataset

dataset = {
    'Student': ['ada', 'chima', 'obi', 'lyon', 'stacy'],
        'rank' : [1, 4, 3, 5, 2],
    'mark' : [95, 70, 80, 60, 90]
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (dataset, index =['RowA', 'RowB', 'RowC', 'RowD', 'RowE'])


# print the student records

print("Student Records\n\n", df)

# Print the size number of elements in the dataframe

print("\nNumber Of Element:\n", df.size)
```

```python
#The pandas DataFrame.T attributes T , transpose rows and columns

# Dataset

dataset = {
    'Student': ['ada', 'chima', 'obi', 'lyon', 'stacy'],
        'rank' : [1, 4, 3, 5, 2],
    'mark' : [95, 70, 80, 60, 90]
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (dataset, index =['RowA', 'RowB', 'RowC', 'RowD', 'RowE'])


# print the student records

print("Student Records\n\n", df)

print("\nTranspose:\n", df.T)
```

```python
# The pandas.DataFrame.head() is used to return the first five rows in our dataset



# Dataset

dataset = {
    'Student': ['ada', 'chima', 'obi', 'lyon', 'stacy', 'Urch'],
        'rank' : [1, 4, 3, 5, 6, 2],
    'mark' : [95, 70, 80, 60, 90, 75]
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (dataset, index =['RowA', 'RowB', 'RowC', 'RowD', 'RowE', 'RowF'])  # Added missing closing quote here


# print the student records

print("Student Records\n\n", df)

print("\nFirst Row :\n", df.head(3))  # Added parentheses to call the head() method)
```

```python
# The pandas.DataFrame.tail() is used to return the last five rows in our dataset


# Dataset

dataset = {
    'Student': ['ada', 'chima', 'obi', 'lyon', 'stacy', 'Urch'],
        'rank' : [1, 4, 3, 5, 6, 2],
    'mark' : [95, 70, 80, 60, 90, 75]
}

# create a DataFrame using the DataFrame() method with index

df = pd.DataFrame (dataset, index =['RowA', 'RowB', 'RowC', 'RowD', 'RowE', 'RowF'])  # Added missing closing quote here


# print the student records

print("Student Records\n\n", df)

print("\nlast three Row :\n", df.tail(3))  # Added parentheses to call the tail() method)
```
