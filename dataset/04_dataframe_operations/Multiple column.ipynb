```python
# Select Multiple columns

''' 
we are going to select more that one column without using built-in-functions.
To select two specific columns from pandas DF mention the column 
you want don't mention column you don't want to display

'''

import pandas as pd

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80]
}

df = pd.DataFrame(data)

print("\nQualifers of Europe Competition\n",df)
print("Table Standing = n", df[['Club','Point']])
```

```python
# Selecting Multiple column in a range you have to mentioned the index numbers in a range separated by a colon ":"

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80],
    'Goal Scored' :[180, 155, 130, 120, 100],
    'Goal conceded' :[45, 50, 30, 25, 30],
    'Number of wins' : [35, 30, 29, 27, 25],
    'Number of loss' : [3, 8, 9, 11, 13],
}


df = pd.DataFrame(data)

print("\nQualifers of Europe Competition =\n" ,df)

# Select columns 3rd to 5th
print("\nSelecting Col Range:\n", df[df.columns[2:5]])

```

```python
''' Adding A New column in pandas

we can add a new columns in the following ways

1. add a new column using the insert()method (set the location)
2. Assign a new column using the assign()method (place at the end)

'''

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80],
    'Goal Scored' :[180, 155, 130, 120, 100],
    'Goal conceded' :[45, 50, 30, 25, 30],
    'Number of wins' : [35, 30, 29, 27, 25],
    'Number of loss' : [3, 8, 9, 11, 13],
}


# display and print the data
df = pd.DataFrame(data)
print("\nQualifers of Europe Competition =\n" ,df)

# insert a new column with insert()
df.insert(loc=4, column ='Foul', value=[9, 12, 10, 15, 14])

#display or show the new inserted column

print("\nUpdate DataFrame:\n", df)
```

```python
#  Assign a new column using the assign()method (place at the end)

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80],
    'Goal Scored' :[180, 155, 130, 120, 100],
    'Goal conceded' :[45, 50, 30, 25, 30],
    'Number of wins' : [35, 30, 29, 27, 25],
    'Number of loss' : [3, 8, 9, 11, 13],
}


# display and print the data
df = pd.DataFrame(data)
print("\nQualifers of Europe Competition =\n" ,df)

# insert a new column with assign() method
df = df.assign(Foul = [9, 12, 10, 15, 14])

# Display the updated DataFrame
print("\nUpdate DataFrame using assign():\n", df)
```
