```python
'''
handling duplicate in pandas means to find and remove duplicates from rows in a DataFrame or Series 
by using the following 
* Duplicated(): find duplicate and 
* drop_duplicate(): Remove Duplicates

'''

# import pandas as pd
import pandas as pd

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc", "Liv", "Manu", "Ars"],
    'Rank':[1, 2, 3, 4, 5, 1, 2, 3],  # Added values for rows 6, 7, 8
    'Point':[100, 95, 90, 85, 80, 100, 95, 90],
    'Goal Scored':[180, 155, 130, 120, 100, 180, 155, 130],  # Added values for rows 6, 7, 8
    'Goal conceded':[45, 50, 30, 25, 30, 45, 50, 30],  # Added values for rows 6, 7, 8
    'Number of wins':[35, 30, 29, 27, 25, 35, 30, 29],  # Added value for row 8
    'Number of loss':[3, 8, 9, 11, 13,3, 8, 9],  # Added value for row 8
}

# create a dataframe with an index
df = pd.DataFrame(data, index=["101", "102", "103", "104", "105", "106", "107", "108"])

# Display the record before cleaning
print("\nQualifer for European Competition =\n", df)

# Find the duplicate
dup = df.duplicated()
print("\nDescribing Duplicates\n", dup)
```

```python
# Remove Duplicate using the drop_duplicate(): To Remove Duplicates

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc", "Liv", "Manu", "Ars"],
    'Rank':[1, 2, 3, 4, 5, 1, 2, 3],  # Added values for rows 6, 7, 8
    'Point':[100, 95, 90, 85, 80, 100, 95, 90],
    'Goal Scored':[180, 155, 130, 120, 100, 180, 155, 130],  # Added values for rows 6, 7, 8
    'Goal conceded':[45, 50, 30, 25, 30, 45, 50, 30],  # Added values for rows 6, 7, 8
    'Number of wins':[35, 30, 29, 27, 25, 35, 30, 29],  # Added value for row 8
    'Number of loss':[3, 8, 9, 11, 13,3, 8, 9],  # Added value for row 8
}

# create a dataframe with an index
df = pd.DataFrame(data, index=["101", "102", "103", "104", "105", "106", "107", "108"])

# Display the record before cleaning
print("\nQualifer for European Competition =\n", df)

# Find the duplicate
dup = df.drop_duplicates()
print("\nTable After Deleting Duplicates\n", dup)
```
