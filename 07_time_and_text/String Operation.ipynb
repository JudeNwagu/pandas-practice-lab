# STRING OPERATIONS ON TEXT  DATA

```python
'''
We can easily perform operations on strings in pandas using the string methods. In this lesson,
we will see how to perform the following string operations on text data in pandas:

1. lower(): Converts text to lowercase
2. upper(): Converts text to uppercase
3. title(): Converts text to title case (capitalizes first letter of each word)
4. len(): Returns the length of each string element
5. count(): Counts the non-empty cells for each column or row
6. contains(): Searches for a specific value in a column and returns boolean results
'''
```

```python
 # Import pandas

import pandas as pd

# lower(): Converts text to lowercase

# Dataset

data =  ["LIVERPOOL", "MANcherster", "ArseNAl", "CHELsea", "BARCA", "REAL madrid"]

# create a series using the series method()

Series = pd.Series(data)

# display the series

print("\nSeries =\n" , Series)

# convert the text to data to lowercase

print("\nlowercase data\n", Series.str.lower())
```

```python
 # upper(): Converts text to uppercase

# Dataset

data =  ["liverpool", "mancherster", "arsenal", "chelsea", "barca", "real madrid"]

# create a series using the series method()

Series = pd.Series(data)

# display the series

print("\nSeries =\n" , Series)

# convert the text to data to lowercase

print("\nUPPERCASE DATA\n", Series.str.upper())

```

```python
# title(): Converts text to title case (capitalizes first letter of each word)

# Dataset

data =  ["liverpool", "mancherster", "arsenal", "chelsea", "barca", "real madrid"]

# create a series using the series method()

Series = pd.Series(data)

# display the series

print("\nSeries =\n" , Series)

# convert the text to data to lowercase

print("\nCapitalize Each First Letter\n", Series.str.title())
```

```python
# len(): Returns the length of each string element


# Dataset

data =  ["liverpool", "mancherster", "arsenal", "chelsea", "barca", "real madrid"]

# create a series using the series method()

Series = pd.Series(data)

# display the series

print("\nSeries =\n" , Series)

# convert the text to data to lowercase

print("\nTotal Letter Number of Football Club\n", Series.str.len())
```

```python
# count(): Counts the non-empty cells for each column or row

# import numpy 
import numpy as np

# Dataset

data =  [np.nan, "liverpool", "mancherster", "arsenal", "chelsea", "barca", "real madrid"]

# create a series using the series method()

Series = pd.Series(data)

# display the series

print("\nSeries =\n" , Series)

# count the series

print("\nTotal Number of Football Club\n", Series.count())
```

```python
#  contains(): Searches for a specific value in a column and returns boolean results

# Dataset

data =  ["liverpool", "mancherster", "arsenal", "chelsea", "barca", "real madrid"]

# create a series using the series method()

Series = pd.Series(data)

# display the series

print("\nSeries =\n" , Series)

# Search for a specific values
print("\n if the value found indicate yes\n", Series.str.contains("chelsea"))
```
