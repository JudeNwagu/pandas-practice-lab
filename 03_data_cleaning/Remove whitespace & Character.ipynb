REMOVE WHITESPACE AND SPECIAL CHARACTER

```python
'''
To remove whitespace or specific character from a text a series or dataframe 
we use the following built in method in python pandas

1. strip(): Removes leading and trailing whitespace or specified characters
2. lstrip(): Removes leading whitespace or specified characters (from the left side)
3. rstrip(): Removes trailing whitespace or specified characters (from the right side)

'''
```

```python
# import pandas 

import pandas as pd

# Data stored in the pandas series as example
data =  ["! liverpool", "mancherster!", "arsenal\t", "chelsea\n\n", "barca \n", " !!real madrid"]

#create a series using the series() method

series = pd.Series(data)

# Display the series

print("Series:\n", series)

# strip the values

print("\nRemove whitespace and special characters:\n",series.str.strip("!\n\n\t\n"))
```

```python
# lstrip() method in python pandas

# lstrip(): Removes leading whitespace or specified characters (from the left side)


# Data stored in the pandas series as example
data =  ["! liverpool", "%mancherster!", "%%arsenal", "chelsea\n\n", "barca \n", " !!real madrid"]

#create a series using the series() method

series = pd.Series(data)

# Display the series

print("Series:\n", series)

# strip the values from the left

print("\nRemove whitespace and special characters From Left:\n",series.str.lstrip("! !! %%"))
```

```python
# rstrip(): Removes leading whitespace or specified characters (from the left side)


# Data stored in the pandas series as example
data =  ["liverpool$$", "!mancherster", "arsenal??", "$chelsea\n\n", "barca \n", " !real madrid"]

#create a series using the series() method

series = pd.Series(data)

# Display the series

print("Series:\n", series)

# strip the values from the left

print("\nRemove whitespace and special characters From Left:\n",series.str.rstrip("! !! %% ? $ \n"))
```
