```python
# import pandas 

import pandas as pd

# dataset

# Dataset

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80],
    'Goal Scored' :[180, 155, 130, 120, 100],
    'Goal conceded' :[45, 50, 30, 25, 30],
    'Number of wins' : [35, 30, 29, 27, 25],
    'Number of loss' : [3, 8, 9, 11, 13],
}

# create a dataframe
df = pd.DataFrame(data, index =["RowA", "RowB", "RowC", "RowD", "RowE"])

# Display
print("\nQualifers of Europe Competition =\n" ,df)

# sort the dataframe in ascending order (default)

print("\nSorting In Ascending Order =\n",df.sort_values(by=['Point']))
```

```python
'''
Sort Dataframe in descending order
To sort the dataframe in descending order, use the sort_values() method. 
set the ascending parameter of the method to false for descending order sort

'''




# Dataset

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80],
    'Goal Scored' :[180, 155, 130, 120, 100],
    'Goal conceded' :[45, 50, 30, 25, 30],
    'Number of wins' : [35, 30, 29, 27, 25],
    'Number of loss' : [3, 8, 9, 11, 13],
}

# create a dataframe
df = pd.DataFrame(data, index =["RowA", "RowB", "RowC", "RowD", "RowE"])

# Display
print("\nQualifers of Europe Competition =\n" ,df)

# sort the dataframe in ascending order (default)

print("\nSorting In Descending Order =\n",df.sort_values(by=['Point'], ascending=False))
```
