```python
# Pandas iterater over rows using the iterrows()

# import pandas

import pandas as pd

# Dataset

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80],
    'Goal Scored' :[180, 155, 130, 120, 100],
    'Goal conceded' :[45, 50, 30, 25, 30],
    'Number of wins' : [35, 30, 29, 27, 25],
    'Number of loss' : [3, 8, 9, 11, 13],
}

# display and print the data
df = pd.DataFrame(data)
print("\nQualifers of Europe Competition =\n" ,df)

# iterate over rows
for row in df.iterrows():
    print(row)
```

```python
'''
ITERTUPLES() to iterate over rows
To iterate over rows, use the python pandas itertuples()method. 
here each row is returned as a python tuple object

'''

# Dataset

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80],
    'Goal Scored' :[180, 155, 130, 120, 100],
    'Goal conceded' :[45, 50, 30, 25, 30],
    'Number of wins' : [35, 30, 29, 27, 25],
    'Number of loss' : [3, 8, 9, 11, 13],
}

# display and print the data
df = pd.DataFrame(data)
print("\nQualifers of Europe Competition =\n" ,df)

# iterate over rows
for row in df.itertuples():
    print(row)
```

```python
''' ITERATE OVER COLUMNS item()method
To iterate each and every column, we use the pandas.items() method
the result will display a label object that is the name of the column and
a column object that is what you have in the column

'''


import pandas as pd

# Dataset

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80],
    'Goal Scored' :[180, 155, 130, 120, 100],
    'Goal conceded' :[45, 50, 30, 25, 30],
    'Number of wins' : [35, 30, 29, 27, 25],
    'Number of loss' : [3, 8, 9, 11, 13],
}

# display and print the data
df = pd.DataFrame(data)
print("\nQualifers of Europe Competition =\n" ,df)

# iterate over columns
for a, b in df.items():
    print(a)
    print(b)
```
