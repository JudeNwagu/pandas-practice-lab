```python
'''
PANDAS Deleting Rows and columns
The drop()method is used in python to delete rows/ columns in a pandas DataFrame
it is used to remove a particular row or column
under the parameter of the drop() method, mention the column you want to delete with the axis.

'''

# import pandas
import pandas as pd

# Dataset

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80],
    'Goal Scored' :[180, 155, 130, 120, 100],
    'Goal conceded' :[45, 50, 30, 25, 30],
    'Number of wins' : [35, 30, 29, 27, 25],
    'Number of loss' : [3, 8, 9, 11, 13],
}

# display and print the data
df = pd.DataFrame(data)
print("\nQualifers of Europe Competition =\n" ,df)

# Drop a column

dpf = df.drop(labels=['Goal conceded'], axis=1)

# Display the updated 

print("\nDataframe After Deleting: \n", dpf)


```

```python
#Dropping Rows using drop()

data = {
    'Club': ["Liv", "Manu", "Ars", "Che", "Manc"],
    'Rank':[1, 2, 3, 4, 5],
    'Point':[100, 95, 90, 85, 80],
    'Goal Scored' :[180, 155, 130, 120, 100],
    'Goal conceded' :[45, 50, 30, 25, 30],
    'Number of wins' : [35, 30, 29, 27, 25],
    'Number of loss' : [3, 8, 9, 11, 13],
}

# display and print the data
df = pd.DataFrame(data)
print("\nQualifers of Europe Competition =\n" ,df)

# Drop a column
# Changed 'axis=index' to 'axis=0' to specify we want to drop rows
dpr = df.drop(labels=[3], axis=0)  

# Display the updated dataframe
print("\nDataframe After Deleting Row: \n", dpr)
```
