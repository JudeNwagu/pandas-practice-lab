```python
''' INDEXING IN PANDAS
 indexing is the selection of specific rows and columns of data from DataFrame.
 A dataframe includes columns, index and data

 indexing in pandas using the indexing operator[]
 indexing in pandas using loc[]
 inexing in pandas usng iloc[]
 '''

# import pandas

import pandas as pd

#Load csv into the dataframe

df = pd.read_csv(r"C:\Users\HP\Documents\sales car inventory.csv")

# Display the csv

print("\nOur Table = \n", df)

# use the indexing operator

indx = df["Color"]

# Result 
print("\n", indx)
```

```python
# Indexing using loc[] attribute

#Load csv into the dataframe

df = pd.read_csv(r"C:\Users\HP\Documents\sales car inventory.csv", index_col = "Model")

# Display the csv

print("\nOur Table = \n", df)

# Retrieve a single row

print("\nUpdated Index = \n", df.loc["Accord"])
```

```python
# Indexing using iloc[] to retrieve the rows and columns by position

#Load csv into the dataframe

df = pd.read_csv(r"C:\Users\HP\Documents\sales car inventory.csv", index_col = "Price")

# Display the csv

print("\nOur Table = \n", df)

# Retrieve a single row

print("\nUpdated Index = \n", df.iloc["2"])
```

```python
# Indexing using iloc[] to retrieve the rows and columns by position

# Load csv into the dataframe without setting index_col initially
# This allows us to check if 'price' column exists before setting it as index
df = pd.read_csv(r"C:\Users\HP\Documents\sales car inventory.csv")

# Check if 'price' column exists in the dataframe
if 'Price' in df.columns:
    # Set 'price' as index if it exists
    df = df.set_index('Price')
else:
    # If 'price' doesn't exist, print available columns
    print("Column 'price' not found. Available columns:", df.columns.tolist())

# Display the csv
print("\nOur Table = \n", df)

# Retrieve a single row using iloc[] with integer position
print("\nUsing iloc with integer position = \n", df.iloc[2])  # This gets the 3rd row (0-indexed)

# If you want to access by the index label (price value), use loc[] instead
# print("\nUsing loc with index label = \n", df.loc[2])  # This gets the row where price=2
```
