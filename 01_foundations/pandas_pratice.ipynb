# To create a pandas dataframe. first we need to create our dataset in using the Dictionary format  'key' & 'value'

```python
# Dataset
# Import pandas library
import pandas as pd


# Dataset
dataset = {
    'student': ['ada', 'chima', 'obi', 'lyon', 'stace'],
    'rank': [1, 2, 3, 4, 5],
    'marks': [90, 80, 70, 85, 91]  # Removed the extra value (92) to match other lists
}

# Create DataFrame with correct capitalization (DataFrame, not dataframe)
df = pd.DataFrame(dataset)

print("student records\n\n", df)
```

```python
# To access group of rows or column in pandas dataframe we used dataframe.loc
# Dataset
import pandas as pd

dataset = {
    'student': ['ada', 'chima', 'obi', 'lyon', 'stace'],
    'rank': [1, 2, 3, 4, 5],
    'marks': [90, 80, 70, 85, 91]  # Removed the extra value (92) to match other lists
}

df = pd.DataFrame(dataset, index = ['RowA', 'RowB', 'RowC', 'RowD', 'RowE'],)
print("student records\n\n", df)

# access the value in the student column corresponding to ROWB

print("\nvalue =", df.loc['RowC','student'])

```

```python
# To access group of rows or column in pandas dataframe we used dataframe.loc
# Dataset
import pandas as pd

dataset = {
    'student': ['ada', 'chima', 'obi', 'lyon', 'stace'],
    'rank': [1, 2, 3, 4, 5],
    'marks': [90, 80, 70, 85, 91] 
}

# Fix 1: Use strings for row labels instead of undefined variables
# Fix 2: Change 'Index' to lowercase 'index' (correct parameter name)
df = pd.DataFrame(dataset, index=['RowA', 'RowB', 'RowC', 'RowD', 'RowE'])
print("student records\n\n", df)

# access the value in the student column corresponding to ROWB
print("\nvalue =", df.loc['RowB','student'])
```

```python
# To access a group of rows by integer position in pandas we used df.iloc
dataset = { 
    'Student' : ['ada', 'chima', 'obi', 'lyon', 'stacy'],
    'Rank' : [1, 4, 3, 5, 2],
    'Mark' : [95, 70, 80, 60, 90]
}

df = pd.DataFrame(dataset, index =['RowA', 'RowB', 'RowC', 'RowD', 'RowE'])

print("student Records\n\n", df)

# Access using rows or columns  by integer positions

print("\nvalue = \n", df.iloc[[1, 2]])
```

```python
# using index argument to set your indexes
dataset = { 
    'Student' : ['ada', 'chima', 'obi', 'lyon', 'stacy'],
    'Rank' : [1, 4, 3, 5, 2],
    'Mark' : [95, 70, 80, 60, 90]
}

# using index argument to set your indexes

df = pd.DataFrame(dataset, index = ['student1', 'student2', 'student3', 'student4', 'student5'])

# print the student record
print("student Records\n\n", df)
                  
```
